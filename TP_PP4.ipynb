### Generacion de columnas temporales (Fecha y Hora)

```python
def split_date_and_time_column(df, column):
    parts = column.split('_')
    if len(parts) >= 3:
        base = '_'.join(parts[:3])
    else:
        base = '_'.join(parts[:2])

    df[[f"{base}_date", f"{base}_time"]] = df[column].str.split(' ', expand=True)

    df[f"{base}_date"] = pd.to_datetime(df[f"{base}_date"])
    df[f"{base}_time"] = pd.to_datetime(df[f"{base}_time"], format='%H:%M:%S').dt.time
    return df

for name, obj in list(globals().items()):
    if name.startswith('df_') and isinstance(obj, pd.DataFrame):
        for col in obj.columns:
            if col.endswith(('_date', '_timestamp', '_at')):
                split_date_and_time_column(obj, col)
```
