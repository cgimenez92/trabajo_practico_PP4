### Generacion de columnas temporales (Fecha y Hora)

```python
import pandas as pd

def split_date_and_time_column(df, column):
    # If already datetime, extract directly
    if pd.api.types.is_datetime64_any_dtype(df[column]):
        df[f"{column}_date"] = df[column].dt.date
        df[f"{column}_time"] = df[column].dt.time
        return df

    # Extract base name
    parts = column.split('_')
    base = '_'.join(parts[:min(3, len(parts))])

    try:
        # Split string into date and time
        date_time_split = df[column].str.split(' ', expand=True)
        df[f"{base}_date"] = pd.to_datetime(date_time_split[0], errors='coerce')
        df[f"{base}_time"] = pd.to_datetime(date_time_split[1], format='%H:%M:%S', errors='coerce').dt.time
    except Exception as e:
        print(f"Error processing column '{column}': {e}")

    return df

for name, obj in list(globals().items()):
    if name.startswith('df_') and isinstance(obj, pd.DataFrame):
        for col in obj.columns:
            if col.endswith(('_date', '_timestamp', '_at')):
                obj = split_date_and_time_column(obj, col)
```
